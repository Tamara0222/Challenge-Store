### Importación de datos



```python

import numpy as np
#mport seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt



```

```python
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.columns
```

#1. Análisis de facturación



```python
tienda.head()
     
```

```python
#Se asignan nombres

tienda['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'

# Calculo de facturación de cada tienda.

fact_tienda = tienda['Precio'].sum()
fact_tienda2 = tienda2['Precio'].sum()
fact_tienda3 = tienda3['Precio'].sum()
fact_tienda4 = tienda4['Precio'].sum()

print(f'la facturación de la tienda 1 es: {fact_tienda}')
print(f'la facturación de la tienda 2 es: {fact_tienda2}')
print(f'la facturación de la tienda 3 es: {fact_tienda3}')
print(f'la facturación de la tienda 4 es: {fact_tienda4}')
```

```python
import matplotlib.pyplot as plt

facturaciones = [fact_tienda, fact_tienda2, fact_tienda3, fact_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

colores = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']

plt.pie(facturaciones, labels=tiendas, colors=colores, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4, 'edgecolor':'white'})

plt.title('Comparativa de facturación por tienda')
plt.show()
```

# 2. Ventas por categoría

```python
#Muestra de por categoria

cat_t1 = tienda['Categoría del Producto'].value_counts()
print("Tienda 1 - Categorías más populares:")
print(cat_t1)

cat_t2 = tienda2['Categoría del Producto'].value_counts()
print("\nTienda 2 - Categorías más populares:")
print(cat_t2)

cat_t3 = tienda3['Categoría del Producto'].value_counts()
print("\nTienda 3 - Categorías más populares:")
print(cat_t3)

cat_t4 = tienda4['Categoría del Producto'].value_counts()
print("\nTienda 4 - Categorías más populares:")
print(cat_t4)
```

```python
# 5 más populares por cada una

cat_t1 = tienda['Categoría del Producto'].value_counts().head(5)
print("Tienda 1 - Categorías más populares:")
print(cat_t1)

cat_t2 = tienda2['Categoría del Producto'].value_counts().head(5)
print("Tienda 2 - Categorías más populares:")
print(cat_t2)

cat_t3 = tienda3['Categoría del Producto'].value_counts().head(5)
print("Tienda 3 - Categorías más populares:")
print(cat_t3)

cat_t4 = tienda4['Categoría del Producto'].value_counts().head(5)
print("Tienda 4 - Categorías más populares:")
print(cat_t4)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

colores_tiendas = [
    ['#FF9999', '#FF6666', '#FFCCCC', '#FFB3B3', '#FF8080'],
    ['#66B3FF', '#3399FF', '#99CCFF', '#66CCFF', '#3399CC'],
    ['#99FF99', '#66FF66', '#CCFFCC', '#66CC66', '#33CC33'],
    ['#FFCC99', '#FFB366', '#FFD9B3', '#FF9933', '#FF8000']
]

categorias = [cat_t1, cat_t2, cat_t3, cat_t4]

# Títulos
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, ax in enumerate(axs.flat):
    ax.pie(categorias[i].values,
           labels=categorias[i].index,
           colors=colores_tiendas[i],
           autopct='%1.1f%%',
           startangle=90,
           wedgeprops={'width':0.4, 'edgecolor':'white'})
    ax.set_title(titulos[i])

plt.suptitle('Top 5 Categorías Más Populares por Tienda', fontsize=16, y=1.03)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd


df_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index = True)

#Promedio de calificacion por tienda

calificacion_promedio = df_tiendas.groupby('tienda')['Calificación'].mean()
print(calificacion_promedio)
```

```python
import matplotlib.pyplot as plt

tiendas = calificacion_promedio.index
promedios = calificacion_promedio.values

colores = ['#66CC99', '#FF9999', '#6699FF', '#FFCC66']

fig, ax = plt.subplots(figsize=(8,6))
barras = ax.bar(tiendas, promedios, color=colores, edgecolor='black', linewidth=1.2)

ax.set_title('Promedio de Calificación por Tienda', fontsize=14, fontweight='bold')
ax.set_ylabel('Calificación Promedio', fontsize=12)
ax.set_ylim(0, 5)

for barra in barras:
    height = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2, height + 0.05, f'{height:.2f}',
            ha='center', va='bottom', fontsize=10, fontweight='bold', color='navy')

plt.xticks(rotation=0, fontsize=11)
plt.tight_layout()
plt.show()

```

# 4. Productos más y menos vendidos

```python


t1 = df_tiendas[df_tiendas['tienda'] == 'Tienda 1']
t2 = df_tiendas[df_tiendas['tienda'] == 'Tienda 2']
t3 = df_tiendas[df_tiendas['tienda'] == 'Tienda 3']
t4 = df_tiendas[df_tiendas['tienda'] == 'Tienda 4']

#Se analiza por tienda

mas_vendidos_t1 = t1['Producto'].value_counts().head(5)
menos_vendidos_t1 = t1['Producto'].value_counts().tail(5)

mas_vendidos_t2 = t2['Producto'].value_counts().head(5)
menos_vendidos_t2 = t2['Producto'].value_counts().tail(5)

mas_vendidos_t3 = t3['Producto'].value_counts().head(5)
menos_vendidos_t3 = t3['Producto'].value_counts().tail(5)

mas_vendidos_t4 = t4['Producto'].value_counts().head(5)
menos_vendidos_t4 = t4['Producto'].value_counts().tail(5)


print("Más vendidos Tienda 1:\n", mas_vendidos_t1)
print("Menos vendidos Tienda 1:\n", menos_vendidos_t1)

print("Más vendidos Tienda 2:\n", mas_vendidos_t2)
print("Menos vendidos Tienda 2:\n", menos_vendidos_t2)

print("Más vendidos Tienda 3:\n", mas_vendidos_t3)
print("Menos vendidos Tienda 3:\n", menos_vendidos_t3)

print("Más vendidos Tienda 4:\n", mas_vendidos_t4)
print("Menos vendidos Tienda 4:\n", menos_vendidos_t4)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4, 2, figsize=(16, 18))  # 4 tiendas, 2 columnas

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos_mas = [mas_vendidos_t1, mas_vendidos_t2, mas_vendidos_t3, mas_vendidos_t4]
datos_menos = [menos_vendidos_t1, menos_vendidos_t2, menos_vendidos_t3, menos_vendidos_t4]

colores_mas = ['#66CC99', '#99FF99', '#33CC33', '#66FF66']
colores_menos = ['#FF9999', '#FF6666', '#FFCCCC', '#FF3333']

for i in range(4):
    # Más vendidos
    barras_mas = axs[i, 0].barh(datos_mas[i].index[::-1], datos_mas[i].values[::-1],
                                color=colores_mas[i], edgecolor='black', linewidth=1)
    axs[i, 0].set_title(f'{tiendas[i]} - Más vendidos', fontsize=12, fontweight='bold')
    axs[i, 0].set_xlabel('Cantidad')
    
    for barra in barras_mas:
        width = barra.get_width()
        axs[i, 0].text(width + 0.5, barra.get_y() + barra.get_height()/2, f'{int(width)}',
                       va='center', ha='left', fontsize=10, color='navy', fontweight='bold')

    # Menos vendidos
    barras_menos = axs[i, 1].barh(datos_menos[i].index, datos_menos[i].values,
                                  color=colores_menos[i], edgecolor='black', linewidth=1)
    axs[i, 1].set_title(f'{tiendas[i]} - Menos vendidos', fontsize=12, fontweight='bold')
    axs[i, 1].set_xlabel('Cantidad')

    for barra in barras_menos:
        width = barra.get_width()
        axs[i, 1].text(width + 0.5, barra.get_y() + barra.get_height()/2, f'{int(width)}',
                       va='center', ha='left', fontsize=10, color='navy', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Más vendidos 
productos_df = pd.concat(
    [mas_vendidos_t1, mas_vendidos_t2, mas_vendidos_t3, mas_vendidos_t4],
    axis=1
)
productos_df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
productos_df.fillna(0, inplace=True)

colores = ['#66CC99', '#FF9999', '#6699FF', '#FFCC66']

ax = productos_df.plot(kind='bar', figsize=(16, 8), color=colores, edgecolor='black', linewidth=1.2)
plt.title('Productos más vendidos por tienda', fontsize=16, fontweight='bold')
plt.ylabel('Cantidad de ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')

for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, color='navy', fontweight='bold')

plt.legend(title='Tiendas')
plt.tight_layout()
plt.show()

#Menos vendidos
productos_df = pd.concat(
    [menos_vendidos_t1, menos_vendidos_t2, menos_vendidos_t3, menos_vendidos_t4],
    axis=1
)
productos_df.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
productos_df.fillna(0, inplace=True)

ax = productos_df.plot(kind='bar', figsize=(16, 8), color=colores, edgecolor='black', linewidth=1.2)
plt.title('Productos menos vendidos por tienda', fontsize=16, fontweight='bold')
plt.ylabel('Cantidad de ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45, ha='right')

for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, color='navy', fontweight='bold')

plt.legend(title='Tiendas')
plt.tight_layout()
plt.show()


```

# 5. Envío promedio por tienda

```python


#envíos
envios_t1 = tienda['Costo de envío'].mean()
print("Envío promedio Tienda 1:",envios_t1)

envios_t2 = tienda2['Costo de envío'].mean()
print("Envío promedio Tienda2:",envios_t2)

envios_t3 = tienda3['Costo de envío'].mean()
print("Envío promedio Tienda 3:",envios_t3)

envios_t4 = tienda4['Costo de envío'].mean()
print("Envío promedio Tienda 4:",envios_t4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

promedios_envio = {
    'Tienda 1': envios_t1,
    'Tienda 2': envios_t2,
    'Tienda 3': envios_t3,
    'Tienda 4': envios_t4
}

df_envios = pd.DataFrame.from_dict(promedios_envio, orient='index', columns=['Costo de envío promedio'])

colores = ['#66CC99', '#FF9999', '#6699FF', '#FFCC66']

ax = df_envios.plot(kind='bar', legend=False, figsize=(8,6),
                    color=colores, edgecolor='black', linewidth=1.2)

plt.title("Costo promedio de envío por tienda", fontsize=14, fontweight='bold')
plt.ylabel("Costo promedio ($)")
plt.xlabel("Tienda")
plt.xticks(rotation=0)

for container in ax.containers:
    ax.bar_label(container, labels=[f"${v:,.2f}" for v in container.datavalues],
                 label_type='edge', fontsize=10, color='navy', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
